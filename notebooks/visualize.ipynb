```python
import json
import os
import sys
import torch
from transformers import AutoModelForCausalLM, AutoTokenizer

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(device)
sys.path.append(os.path.join(os.getcwd(),"../libraries/repeng"))

from repeng import ControlVector, ControlModel, DatasetEntry
```

```python
control_vector = torch.load("../nlp/data/vectors/Elaborate_Simple.pt")
```

```python
print(control_vector)

cv_data = []

for i in control_vector.directions:
    print(control_vector.directions[i])
    cv_data.append(control_vector.directions[i])
    print(len(control_vector.directions[i]))
```

```python
import numpy as np
import matplotlib.pyplot as plt


plt.figure(figsize=(10, 6))  # Adjust the figure size as needed
plt.imshow(cv_data, cmap='seismic', aspect='auto')
plt.colorbar(label='Values')  # Add color bar with label
plt.title('Heatmap of 31 Arrays')  # Add title
plt.xlabel('Index')  # Add x-axis label
plt.ylabel('Array')  # Add y-axis label
plt.show()
```

```python
mean_values = np.mean(cv_data, axis=0)
```

```python
plt.figure(figsize=(10, 1))  # Adjust the figure size as needed
plt.imshow([mean_values], cmap='seismic', aspect='auto', extent=[0, len(mean_values), 0, 1])
plt.colorbar(label='Mean Value')  # Add color bar with label
plt.title('Average Control Vector Across Layers - Elaborate vs Simple')  # Add title
plt.xlabel('Index X')  # Add x-axis label
plt.ylabel('Intensity')  # Add y-axis label
plt.show()
```

```python
mean_values_trans = np.mean(cv_data, axis=1)

plt.figure(figsize=(10, 1))  # Adjust the figure size as needed
plt.imshow([mean_values_trans], cmap='seismic', aspect='auto', extent=[0, len(mean_values_trans), 0, 1])
plt.colorbar(label='Mean Value')  # Add color bar with label
plt.title('Average CV Weight Magnitude for Each Layer  - Elaborate vs Simple')  # Add title
plt.xlabel('Index X')  # Add x-axis label
plt.ylabel('Intensity')  # Add y-axis label
plt.show()
```
